```python
b = open('file11.csv','w')
b.write("name,c,oop,ir,daa,dsa,\n")
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    c = int(input("Enter the marks of c = "))
    oop = int(input("Enter the marks of oop = "))
    ir = int(input("Enter the marks of ir = "))
    daa = int(input("Enter marks of daa = "))
    dsa = int(input("Enter the marks of dsa = "))

    
   
    info = f"{name},{c},{oop},{ir},{daa},{dsa}\n"
    b.write(info)
b.close()
```
